# Question #1
Use python and Jupyter to create a plot of $\sin(\pi x)$ and $\cos(\pi x)$ over the interval $x\in [-1, 1]$. Both functions should be plotted in the same graph. Use a legend to title each curve. Make the color of the first curve black. Make the second curve a dashed line and color it purple.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-1, 1, 100)
y1 = np.sin(np.pi*x)
y2 = np.cos(np.pi*x)
plt.figure(figsize=(8,6))
plt.plot(x, y1, label=r'$\sin(\pi x)$', color='black')
plt.plot(x, y2, label=r'$\cos(\pi x)$', linestyle='--', color='purple')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.title(r'$\sin(\pi x)$ and $\cos(\pi x)$')
plt.legend()
plt.show()
```

# Question #2
Use python and Jupyter to create a plot of $\exp(-x^2)$ over the interval $x\in [-4, 4]$. Create a second graph of the same function, but put the y-axis on a log scale.


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4, 4, 100)
y1 = np.exp(-x**2)
y2 = np.exp(-x**2)
plt.figure(figsize=(8,6))
plt.plot(x, y1, label=r'$\exp(-x^2)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.title(r'$\exp(-x^2)$')
plt.legend()
plt.show()

#on a logscale
plt.figure(figsize=(8,6))
plt.semilogy(x, y2, label=r'$\exp(-x^2)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$ (log scale)')
plt.title(r'$\exp(-x^2)$ on log scale')
plt.legend()
plt.show()
```
