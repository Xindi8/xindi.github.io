# Question #1
Use python and Jupyter to create a plot of $\sin(\pi x)$ and $\cos(\pi x)$ over the interval $x\in [-1, 1]$. Both functions should be plotted in the same graph. Use a legend to title each curve. Make the color of the first curve black. Make the second curve a dashed line and color it purple.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the interval
x = np.linspace(-1, 1, 500)

# Define the functions
y1 = np.sin(np.pi * x)
y2 = np.cos(np.pi * x)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y1, color='black', label='sin(x)')
plt.plot(x, y2, linestyle='--', color='purple', label='cos(x)')

# Add legend, labels, and grid
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of sin(x) and cos(x) over x  [-1, 1]")
plt.grid(True)
plt.show()
```
